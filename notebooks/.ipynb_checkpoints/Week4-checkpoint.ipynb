```python
#   
from textblob import TextBlob
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import nltk

#   punkt   TextBlob
nltk.download('punkt')

#      Project Gutenberg
url = 'https://www.gutenberg.org/files/514/514-0.txt'  # URL   "Little Women"
response = requests.get(url)

#    
with open('little_women.txt', 'w', encoding='utf-8') as file:
    file.write(response.text)

#    
with open('little_women.txt', 'r', encoding='utf-8') as file:
    text = file.read()

#   TextBlob   
blob = TextBlob(text)

# 1.      
sentences = blob.sentences  #    
words = blob.words          #    

#     
print(f" : {len(sentences)}")
print(f" : {len(words)}")

# 2.   
#    
sentiment = blob.sentiment
print(f"  (, ): {sentiment}")

#    5 
for sentence in sentences[:5]:
    print(sentence)
    print(f": {sentence.sentiment}")

# 3.    (POS-tagging)
#      
pos_tags = blob.tags
print(f"   20 : {pos_tags[:20]}")

# 4.   
#           
word_freq = Counter(blob.words.lower())  #   
common_words = word_freq.most_common(10)  #  10   
print(f"10   : {common_words}")

# 5.   
#     
words, counts = zip(*common_words)  #       

plt.figure(figsize=(10, 6))
sns.barplot(x=list(words), y=list(counts))
plt.title(' 10     "Little Women"')
plt.show()
```
