```python
import pandas as pd
import os

file_path = 'data/students_lunch.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    
    print(df.head())

    rich_students = df[df['Сумма на обед'] > 500]
    print("\nСтуденты, которые тратят на обед больше 500:")
    print(rich_students)
else:
    print(f"Файл {file_path} не найден.")
```

```python
import pandas as pd
import os

file_path = 'data/students_lunch.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    

students = pd.DataFrame(df)

students_with_A = students[students['Имя студента'].str.startswith('А')]

print(students_with_A)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Фамилия': ['Иванов', 'Малик', 'Сидоров', 'Кузнецов', 'Смирнов'],
        'Имя': ['Алексей', 'Федр', 'Сергей', 'Ольга', 'Ирина'],
        'Математика': [85, 78, 92, 88, 75],
        'История': [90, 85, 88, 76, 80]}

students = pd.DataFrame(data)

students.to_csv('data/students.csv', index=False)

students.plot(x='Фамилия', y=['Математика', 'История'], kind='bar')
plt.title("Оценки по Математике и Истории")
plt.ylabel("Оценка")
plt.show()
```
