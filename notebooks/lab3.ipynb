```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

hours_of_study = np.array([2, 3, 4, 6, 7, 10, 11]).reshape(-1, 1)  # reshape(-1, 1) для формирования 2D массива
exam_scores = np.array([50, 55, 65, 75, 80, 90, 95])

model = LinearRegression()
model.fit(hours_of_study, exam_scores)

predicted_scores = model.predict(hours_of_study)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Уравнение линейной регрессии: успеваемость = {slope:.2f} * время подготовки + {intercept:.2f}")

plt.scatter(hours_of_study, exam_scores, color='blue', label='Фактические данные')
plt.plot(hours_of_study, predicted_scores, color='red', label='Линейная регрессия')

plt.title('Влияние времени подготовки на успеваемость')
plt.xlabel('Часы подготовки')
plt.ylabel('Успеваемость (%)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

experience = np.array([1, 3, 5, 7, 9, 10]).reshape(-1, 1) 
salary = np.array([20, 35, 45, 50, 60, 75])  

model = LinearRegression()
model.fit(experience, salary)

predicted_salary = model.predict(experience)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Уравнение линейной регрессии: зарплата = {slope:.2f} * стаж + {intercept:.2f}")

plt.scatter(experience, salary, color='blue', label='Фактические данные')
plt.plot(experience, predicted_salary, color='red', label='Линейная регрессия')

plt.title('Влияние стажа работы на заработную плату')
plt.xlabel('Стаж работы (лет)')
plt.ylabel('Заработная плата (тыс. руб.)')
plt.legend()
plt.show()
```
