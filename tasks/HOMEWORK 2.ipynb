```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    '': ['', '', '', ''],
    ', ': [-5, 0, -10, -15],
    ', ': [30, 35, 25, 20],
    ',  ()': [50, 40, 60, 70],
    ',  ()': [10, 5, 20, 30]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.plot(df[''], df[', '], label='', marker='o')
plt.plot(df[''], df[', '], label='', marker='o')

plt.title('    (C)')
plt.xlabel('')
plt.ylabel(' (C)')
plt.legend()

plt.show()

plt.figure(figsize=(10, 5))
plt.bar(df[''], df[',  ()'], width=0.4, label=' ', color='blue', align='center')
plt.bar(df[''], df[',  ()'], width=0.4, label=' ', color='green', align='edge')

plt.title('    ()')
plt.xlabel('')
plt.ylabel('  ()')
plt.legend()

plt.show()
```
