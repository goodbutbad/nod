```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Город': ['Бишкек', 'Ош', 'Каракол', 'Нарын'],
    'Температура, Январь': [-5, 0, -10, -15],
    'Температура, Июль': [30, 35, 25, 20],
    'Осадки, Январь (мм)': [50, 40, 60, 70],
    'Осадки, Июль (мм)': [10, 5, 20, 30]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.plot(df['Город'], df['Температура, Январь'], label='Январь', marker='o')
plt.plot(df['Город'], df['Температура, Июль'], label='Июль', marker='o')

plt.title('Температура в городах Кыргызстана (°C)')
plt.xlabel('Города')
plt.ylabel('Температура (°C)')
plt.legend()

plt.show()

plt.figure(figsize=(10, 5))
plt.bar(df['Город'], df['Осадки, Январь (мм)'], width=0.4, label='Осадки Январь', color='blue', align='center')
plt.bar(df['Город'], df['Осадки, Июль (мм)'], width=0.4, label='Осадки Июль', color='green', align='edge')

plt.title('Осадки в городах Кыргызстана (мм)')
plt.xlabel('Города')
plt.ylabel('Количество осадков (мм)')
plt.legend()

plt.show()
```
